```python
from deluge_client import DelugeRPCClient
```

```python
client = DelugeRPCClient('192.168.1.2', 58846, 'michael', 'password')
```

```python
client.connect()
```

```python
client.connected
```

```python
tracker_statuses = client.call('core.get_torrents_status', {}, ['name','tracker_status'])
tracker_statuses
```

```python
len(tracker_statuses)
```

```python
errors = [ts for ts in tracker_statuses if b"Error: Unregistered torrent" in tracker_statuses[ts][b'tracker_status']]
errors
```

```python
len(errors)
```

```python
for error in errors:
    print(error)
    remove = client.call('core.remove_torrent', error, True)
    print(remove)
```
